```python
# Best way to load 
pic = Image.open('../DATA/00-puppy.jpg')
```

```python
pic_arr = np.asarray(pic)
plt.imshow(pic_arr)
```

```python
plt.subplot(221)
plt.imshow(pic_arr[:,:,0]) # Red Channel
plt.subplot(222)
plt.imshow(pic_arr[:,:,1]) # Blue Channel
plt.subplot(223)
plt.imshow(pic_arr[:,:,2]) # Green Channel
```

```python
fig, axs = plt.subplots(2,2, figsize = (10,8))
axs[0,0].imshow(pic_arr)
axs[0,0].title.set_text("Orginal Image")
axs[0,1].imshow(pic_arr[:,:,0])
axs[0,1].title.set_text("Red Channel")
axs[1,0].imshow(pic_arr[:,:,1])
axs[1,0].title.set_text("Green Channel")
axs[1,1].imshow(pic_arr[:,:,2])
axs[1,1].title.set_text("Blue Channel")
```

```python
fig, axs = plt.subplots(2,2, figsize = (10,8))
axs[0,0].imshow(pic_arr)
axs[0,0].title.set_text("Orginal Image")
axs[0,1].imshow(pic_arr[:,:,0], cmap = 'gray')
axs[0,1].title.set_text("Red Channel")
axs[1,0].imshow(pic_arr[:,:,1], cmap = 'gray')
axs[1,0].title.set_text("Green Channel")
axs[1,1].imshow(pic_arr[:,:,2], cmap = 'gray')
axs[1,1].title.set_text("Blue Channel")
```
